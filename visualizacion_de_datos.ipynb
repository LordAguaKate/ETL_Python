# Visualización de datos con Python
Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df = pd.read_csv('inmigrantes_canada.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dic = {'Año':colombia.index, 'Inmigrantes':colombia.values.tolist()}
col_dic
```

```python
datos_col = pd.DataFrame(col_dic)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Inmigración de Colombia a Canadá (1980-2013)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Inmigración de Colombia a Canadá (1980-2013)')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de Colombia a Canadá\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,4))
ax[0].plot(datos_col['Año'], datos_col['Inmigrantes'], color='green')
ax[0].set_title('Inmigración de Colombia a Canadá\n1980 - 2013')
ax[0].set_xlabel('Año')
ax[0].set_ylabel('Número de Inmigrantes')
ax[0].xaxis.set_major_locator(plt.MultipleLocator(5))
ax[0].grid()

ax[1].boxplot(datos_col['Inmigrantes'])
ax[1].set_title('Distribución de Inmigrantes de Colombia a Canadá\n1980 - 2013')
ax[1].set_xlabel('Colombia')
ax[1].set_ylabel('Número de Inmigrantes')
ax[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

axs[0, 0].plot(df.loc['Colombia', anos])
axs[0, 0].set_title('Colombia')

axs[0, 1].plot(df.loc['Brasil', anos])
axs[0, 1].set_title('Brasil')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Perú', anos])
axs[1, 1].set_title('Perú')

for ax in axs.flat:
    ax.set_xlabel('Año')
    ax.set_ylabel('Número de Inmigrantes')
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid()

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Inmigración de Países Sudamericanos a Canadá\n1980-2013', fontsize=16)

axs[0, 0].plot(df.loc['Colombia', anos])
axs[0, 0].set_title('Colombia')

axs[0, 1].plot(df.loc['Brasil', anos])
axs[0, 1].set_title('Brasil')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Perú', anos])
axs[1, 1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

for ax in axs.flat:
    ax.set_xlabel('Año')
    ax.set_ylabel('Número de Inmigrantes')
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'], lw=1, marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de Colombia a Canadá\n1980 - 2013', fontsize=18)
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'], lw=1, marker='o', color='r')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de Colombia a Canadá\n1980 - 2013', fontsize=18)
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--', linewidth=0.5)
plt.show()
```

```python
sudamerica = df.query('Region == "América del Sur"')
sudamerica.head(10)
```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive',
            'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))

ax.bar(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigración Total de Países Sudamericanos a Canadá\n1980 - 2013', fontsize=16)
ax.set_xlabel('')
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive',
            'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigración Total de Países Sudamericanos a Canadá\n1980 - 2013', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values(by='Total', ascending=True)

colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive',
            'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración Total de Países Sudamericanos a Canadá\n1980 - 2013', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

fig.savefig('inmigracion_paises_canada.png', transparent=False ,dpi=600, bbox_inches='tight')

plt.show()
```

```python
colores = []

for pais in sudamerica_sorted.index:
    if pais == 'Colombia':
        colores.append('forestgreen')
    else:
        colores.append('silver')

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración Total de Colombianos a Canadá\n1980 - 2013', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = []

for pais in sudamerica_sorted.index:
    if pais == 'Colombia':
        colores.append('forestgreen')
    else:
        colores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración Total de Colombianos a Canadá\n1980 - 2013', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i,j in enumerate(sudamerica_sorted['Total']):
    ax.text(j + 20, i, str(j), color='black', va='center', fontsize=10, ha='left')

plt.show()
```

```python
colores = []

for pais in sudamerica_sorted.index:
    if pais == 'Colombia':
        colores.append('forestgreen')
    else:
        colores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='y', which='both', size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración Total de Colombianos a Canadá\n1980 - 2013', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i,j in enumerate(sudamerica_sorted['Total']):
    ax.text(j + 20, i, str(j), color='black', va='center', fontsize=10, ha='left')

fig.savefig('inmigracion_colombia_canada1980_2013.png', transparent=False ,dpi=600, bbox_inches='tight')

plt.show()
```
