# Visualización de datos con Python
Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df = pd.read_csv('inmigrantes_canada.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dic = {'Año':colombia.index, 'Inmigrantes':colombia.values.tolist()}
col_dic
```

```python
datos_col = pd.DataFrame(col_dic)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Inmigración de Colombia a Canadá (1980-2013)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Inmigración de Colombia a Canadá (1980-2013)')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.show()
```
